```python
import pandas as pd


# Tüm teşhisleri küçük harfe çevirip boşluklardan arındır (aynı teşhislerin farklı yazımlarını engellemek için)
df["diagnosis"] = df["diagnosis"].str.lower().str.strip()

# Tüm teşhisleri ve tekrar sayılarını al
diagnosis_counts = df["diagnosis"].value_counts()

# Tüm teşhisleri ve tekrar sayılarını yazdır
for diagnosis, count in diagnosis_counts.items():
    print(f"{diagnosis}: {count} kez")

# Teşhislerin tekrar sayılarını CSV dosyasına yaz 
diagnosis_counts.to_csv("diagnosis_frequencies.csv", header=["count"])

```
